```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_community.vectorstores import Neo4jVector
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda, RunnablePassthrough
from langchain_core.prompts import PromptTemplate
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.llms import HuggingFacePipeline
from langchain_core.messages import AIMessage  # AIMessage 임포트 추가
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
from unsloth import FastLanguageModel
import os
from langchain_community.graphs import Neo4jGraph
from langchain.chains import GraphCypherQAChain
from neo4j import GraphDatabase
from neo4j_graphrag.retrievers import Text2CypherRetriever
import logging

logging.basicConfig()
logging.getLogger("langchain.graph").setLevel(logging.INFO)

# 연결 설정
uri = 'neo4j+s://d7439c49.databases.neo4j.io'
user = 'neo4j'
password = 'EMM7ZD6VbuuEZLA1xeNE4wvoR1Scmf6W87bVHFYx1TE'

# Neo4j 그래프 초기화
graph = Neo4jGraph(
    url=uri,
    username=user,
    password=password
)

driver = GraphDatabase.driver(uri, auth=(user, password))

print(graph.schema)
query_result = []

def run_query(func):
    with driver.session() as session:
        session.read_transaction(func)

def get_symetric_key(tx):
    query = """
        MATCH (a {명칭:'대칭키암호'}) - [r:종류] -> (b)
        RETURN b.명칭 AS 명칭, b.키길이 AS 키길이
        """
    result = tx.run(query)
    for record in result:
        query_result.append('{}(키길이:{})'.format(record['명칭'], record['키길이']))

def get_context(question: str) -> str:
    # 질문은 현재 쿼리에 직접 반영되지 않음, 필요 시 동적 쿼리 생성 로직 추가
    run_query(get_symetric_key)
    return ' '.join(query_result)


# 단계 7: 언어모델(LLM) 생성
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="beomi/Llama-3-Open-Ko-8B-Instruct-preview",
    max_seq_length=16384,
    dtype=None,
    load_in_4bit=True,
)

llm_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128,
    eos_token_id=tokenizer.eos_token_id
)

llm = HuggingFacePipeline(pipeline=llm_pipeline)

prompt = PromptTemplate.from_template(
    """다음 컨텍스트를 사용하여 질문에 답변하세요.
    답변은 간결하고 정확해야 하며, 한국어로 작성하세요.
    컨텍스트에 없는 정보는 사용하지 마세요.

#Context:
{context}

#Question:
{question}

#Answer:"""
)


# 체인 구성
chain = (
    {
        "context": RunnableLambda(get_context),
        "question": RunnablePassthrough()
    }
    | prompt
    | llm
    | StrOutputParser()
)


```

```python
question = "대칭키 암호 종류는?"
response = chain.invoke(question)
print(response)
```
